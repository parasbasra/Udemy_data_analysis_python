# Udemy Data Analysis and Visualisation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.show()
```

## Importing the dataset

```python
data = pd.read_csv(r'C:\Users\Lenovo\Downloads\udemy_dataset.csv')
```

```python
data.isnull().sum()
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
```

## After cleaning the data , it has 3676 rows and 11 columns

```python
data
```

## 1. Distribution of free and paid courses across different levels and subjects.

```python
data['is_paid'] = data['is_paid'].replace({True: 'Paid', False: 'Free'})
```

```python
# Distribution of paid and free courses
sns.countplot(x='is_paid', data=data)
plt.title('Distribution of Paid and Free Courses')
plt.show()

# Distribution of courses across levels
plt.figure(figsize=(10,6))
sns.countplot(x='level', data=data, order=data['level'].value_counts().index)
plt.title('Distribution of Courses Across Levels')
plt.xticks(rotation=0)
plt.show()

# Distribution of courses across subjects
plt.figure(figsize=(10,6))
sns.countplot(x='subject', data=data, order=data['subject'].value_counts().index)
plt.title('Distribution of Courses Across Subjects')
plt.xticks(rotation=0)
plt.show()
```

## 2. Price Analysis for paid and free courses:

```python
# Distribution of prices for paid courses
plt.figure(figsize=(12,6))
sns.histplot(data[data['is_paid'] == True]['price'], bins=20, kde=True)
plt.title('Distribution of Prices for Paid Courses')
plt.xlabel('Price')
plt.xticks(fontsize=7)
plt.ylabel('Frequency')
plt.show()
```

## 3. Subscribers and Reviews:

```python
# Scatter plot of subscribers vs. reviews
sns.scatterplot(x='num_subscribers', y='num_reviews', data=data)
plt.title('Subscribers vs. Reviews')
plt.xlabel('Number of Subscribers')
plt.ylabel('Number of Reviews')
plt.show()
```

## 4. Time Trends

```python
# Convert 'published_timestamp' to datetime
data['published_timestamp'] = pd.to_datetime(data['published_timestamp'])

# Distribution of courses over time
data['published_year'] = data['published_timestamp'].dt.strftime('%Y')
plt.figure(figsize=(10, 6))
sns.countplot(x='published_year', data=data, order=data['published_year'].value_counts().sort_index().index)
plt.title('Distribution of Courses Over Time')
plt.xticks(rotation=0)
plt.show()
```

## 5. Top Courses with highest subscribers.

```python
# Top courses by number of subscribers
top_courses = data.nlargest(10, 'num_subscribers')
sns.barplot(x='num_subscribers', y='course_title', data=top_courses)
plt.title('Top 10 Courses by Subscribers')
plt.xlabel('Number of Subscribers')
plt.ylabel('Course Title')
plt.show()
```

## 6. Paid vs. Free Courses Subscriber Comparison:

```python
data[['Duration','Unit']] = data.content_duration.str.split(' ',expand=True)
```

```python
# Boxplot comparing subscribers for paid and free courses
sns.boxplot(x='is_paid', y='num_subscribers', data=data)
plt.title('Paid vs. Free Courses Subscriber Comparison')
plt.xlabel('Is Paid')
plt.ylabel('Number of Subscribers')
plt.show()
```

## 7. Level-wise Distribution for Each Subject:

```python
# Level-wise distribution for each subject
plt.figure(figsize=(10,6))
sns.countplot(x='subject', hue='level', data=data, order=data['subject'].value_counts().index)
plt.title('Level-wise Distribution for Each Subject')
plt.xticks(rotation=0)
plt.show()
```

## 8. Price vs. Subscribers:

```python
# Scatter plot of price vs. subscribers
plt.figure(figsize=(12,6))
sns.scatterplot(x='price', y='num_subscribers', data=data)
plt.title('Price vs. Subscribers')
plt.xlabel('Price')
plt.xticks(fontsize=7)
plt.ylabel('Number of Subscribers')
plt.show()
```

## 9. Reviews vs. Subscribers for Top Subjects:

```python
# Select top subjects based on the number of courses
top_subjects = data['subject'].value_counts().head(3).index

# Scatter plot of reviews vs. subscribers for top subjects
sns.scatterplot(x='num_subscribers', y='num_reviews', hue='subject', data=data[data['subject'].isin(top_subjects)])
plt.title('Reviews vs. Subscribers for Top Subjects')
plt.xlabel('Number of Subscribers')
plt.ylabel('Number of Reviews')
plt.legend(title='Subject')
plt.show()
```

## 10. Lecture Distribution by Level:

```python
# Boxplot of lectures by level
plt.figure(figsize=(10,6))
sns.boxplot(x='level', y='num_lectures', data=data, order=data['level'].value_counts().index)
plt.title('Lecture Distribution by Level')
plt.xlabel('Level')
plt.ylabel('Number of Lectures')
plt.xticks(rotation=0)
plt.show()
```

## 11. Content Duration Distribution:

```python
# Distribution of content duration
sns.histplot(data['content_duration'].apply(lambda x: float(x.split()[0])), bins=20, kde=True)
plt.title('Distribution of Content Duration')
plt.xlabel('Content Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```
